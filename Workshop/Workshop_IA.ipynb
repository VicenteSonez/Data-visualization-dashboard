```python
import pandas as pd

# Cargar dataset principal
df = pd.read_csv('../Data/EGRE_DATOS_ABIERTOS_2020.csv', delimiter=';', encoding='latin-1', low_memory=False)

# Cargar diccionario de enfermedades
diseases = pd.read_excel('../Data/Diccionario BD egresos hospitalario.xlsx', sheet_name='codigo CIE-10', header=8)

# Realizar el merge (cruce) entre el diagnostico principal (DIAG1) y el codigo de subcategoria
df = pd.merge(df, diseases[['CODIGO SUBCATEGORIA', 'GLOSA SUBCATEGORIA']], left_on='DIAG1', right_on='CODIGO SUBCATEGORIA', how='left')

# Renombrar la columna para identificarla como el nombre de la enfermedad
df = df.rename(columns={'GLOSA SUBCATEGORIA': 'NOMBRE_ENFERMEDAD'})

# Opcional: Eliminar la columna 'CODIGO SUBCATEGORIA' ya que es igual a 'DIAG1'
df = df.drop(columns=['CODIGO SUBCATEGORIA'])

df.head()
```
